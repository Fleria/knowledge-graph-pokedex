Simple UI using PySide

```python
from llm_integration import question_to_cypher_and_answer
```

```python
import sys
from PySide6.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QHBoxLayout,
    QLineEdit, QPushButton, QTextEdit, QLabel
)
from PySide6.QtGui import QIcon
from PySide6.QtCore import Qt, QSize, QTimer


class PokedexApp(QWidget):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("Pokédex")
        self.setStyleSheet("background-color: #d62828;")
        self.setWindowIcon(QIcon("rotom.ico"))

        # layout
        main_layout = QVBoxLayout()
        input_layout = QHBoxLayout()

        # input field
        self.entry = QLineEdit()
        self.entry.setPlaceholderText("Ask me a question!")
        self.entry.setFixedHeight(32)
        self.entry.setStyleSheet("""
            QLineEdit {
                font-family: Consolas;
                font-size: 14px;
                border: 2px solid #333;
                border-radius: 5px;
                padding: 4px;
            }
        """)
        self.entry.returnPressed.connect(self.ask_question) #enable pressing enter to ask question

        circle = QLabel()
        circle.setFixedSize(QSize(20, 20))  # diameter
        circle.setStyleSheet("""
            QLabel {
                background-color: blue;
                border-radius: 10px;
            }
        """)

        input_layout.addWidget(circle)
        input_layout.addWidget(self.entry)

        # button
        self.ask_button = QPushButton("Go")
        self.ask_button.setStyleSheet("""
            QPushButton {
                background-color: #f7e6a8;
                color: black;
                font-family: Consolas;
                font-weight: bold;
                font-size: 14px;
                border-radius: 6px;
                padding: 6px 12px;
            }
            QPushButton:hover {
                background-color: #d9d34d;
            }
        """)
        self.ask_button.clicked.connect(self.ask_question)
        input_layout.addWidget(self.ask_button)

        # output box
        self.output_box = QTextEdit()
        self.output_box.setReadOnly(True)
        self.output_box.setStyleSheet("""
            QTextEdit {
                background-color: #078211;
                color: #000000;
                font-family: "Courier New", monospace;
                font-size: 14px;
                border: 4px solid #00ff66;
                border-radius: 6px;
                padding: 8px;
                text-shadow: 0 0 8px #00ff66; /* neon glow effect */
            }
        """)

        main_layout.addLayout(input_layout)
        main_layout.addWidget(self.output_box)
        self.setLayout(main_layout)

    def ask_question(self):
        user_input = self.entry.text().strip()
        if not user_input:
            return
        response = question_to_cypher_and_answer(user_input)
        self.output_box.append(f"> {user_input}\n{response}\n")
        self.entry.clear()


app = QApplication.instance()
if not app:
    app = QApplication([])

window = PokedexApp()
window.resize(600, 500)
window.show()
```
